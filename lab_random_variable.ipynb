Use the numerical variables from our dataset this week.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

numerical = pd.read_csv('numerical_data.csv')
```

```python
numerical.head()
```

Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
for col in numerical.columns:
    plt.figure(figsize=(6, 4))
    numerical[col].hist(bins=50)
    plt.title('Distribution of' + col)
    plt.xlabel(col)
    plt.ylabel('Distribution')
    plt.tight_layout()
    plt.show()
```

```python
customer_lifetime_value: exp distribution
income: uniform with outliers?
monthly_premium_auto: exp
months_since_last_claim: no standard bistribution? Do not find any pattern
months_since_policy_inception: normal
number_of_open_complaints: discrete - exp.
number_of_policies: discrete - exp.
total_claim_amount: exp.
Year: not applicable (only one value)
Month: not applicable 
Day: uniform
```

Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
import numpy as np

df_log_transformed = np.log1p(numerical)

for col in df_log_transformed.columns:
    plt.figure(figsize=(6, 4))
    df_log_transformed[col].hist(bins=50)
    plt.title('Log transformed distribution of' + col)
    plt.xlabel(col)
    plt.ylabel('Log transformed distribution')
    plt.tight_layout()
    plt.show()
```
